## New type variable: Tuples with ()
What is the difference between List and Tuple? 
- List is mutable and Tuples are immutable
- Immutable means we cannot change it, once we created it, it is fixed anymore

```python
aTuple = (3,4,3)
aList = [3,4,3]

print(aTuple)
print(aList)

aList[0] = 5 # mutable
# aTuple[0] = 5 # gives error

# but you can assign whole new Tuple
aTuple = (5,'hi', '5')
print(aTuple)

# we can convert tuples into list
newL = list(aTuple)
print(type(newL))
# once we changed it into list, we can change its elements
newL[0] = 'hello list'
print(newL)
```

Why we have both tuples nad list? why we need them?
- We can think of a tuple as being like a list that has a security measure on it. It has a lock on it. This would be if the information is really important and you don't want it to be modified by accident, maybe this is some original data and you want to make sure that the original data are never modified so you can store the original data as a tuple.
- Then when you need to grab a little bit of data, you need to select a little bit of data. Then you could select that piece of data and turn that into a list so you could continue working with it and changing it or running analysis on it.

The difference in how you can call certain NumPy functions like `mean()` and `square()` is due to how they are implemented in NumPy.

1. **Direct method call (`a.mean()`)**: Some functions in NumPy are implemented as methods directly on NumPy arrays. These methods are designed to be convenient for common operations and often provide more functionality or customization options compared to their counterparts in the `np` namespace. For example, `mean()` is implemented as a method of NumPy arrays, allowing you to directly call `a.mean()` to compute the mean of the array `a`.

2. **Using `np` namespace (`np.square(a)`)**: Other functions in NumPy are implemented as standalone functions in the NumPy namespace (`np`). These functions are designed to work with various types of inputs, not just NumPy arrays. Therefore, you need to explicitly specify the input as an argument to the function when using it. For example, `np.square()` is a standalone function that computes the element-wise square of an array. To use it, you need to pass the array `a` as an argument: `np.square(a)`.

To determine which approach to use, you can consult the NumPy documentation or use auto-completion in your integrated development environment (IDE) or text editor. Typically, functions implemented as methods of NumPy arrays (like `mean()`, `sum()`, `max()`, etc.) are designed for convenience and readability when working directly with arrays, while functions in the `np` namespace are more general-purpose and can accept a wider range of inputs.

# Broadcasting in Numpy
Important feature for Data Science and Linear Algebra

```python
# question: how do we add W matrix to each row of X seperately?
# ----------------------------------------------------
# first solution
X = np.array([ [1,2,3],
             [2,3,4],
             [3,4,5],
             [4,5,6]  ])

w = np.array([10,20,30])

Y = X
Y = X + w # assigning to a new variable

print(X)
print(Y)

# ----------------------------------------------------
# second solution
X2 = np.array([ [1,2,3],
             [2,3,4],
             [3,4,5],
             [4,5,6]  ])

w2 = np.array([10,20,30])

# question: how do we add W matrix to each row of X seperately?
Y2 = X2

for i in range((X2).shape[0]):
    Y2[i,:] = X2[i,:] + w2 # we are not assigning Y to new pointer, we just change its elements values

print('')
print('printing other solution')
print(X2)
print(Y2)
```

In the above code, (in first solution) you're actually re-assigning Y to be a different value, which means it becomes a different thing and therefore requires a different pointer. That's different from replacing individual values in a list or numpy array (in the second solution).

The key difference is whether the entire variable is replaced (new location in memory) vs whether an element of the variable is changed (same location in memory).

```python
import copy

X = np.array([ [1,2,3],
             [2,3,4],
             [3,4,5],
             [4,5,6]  ])

w = np.array([10,20,30])

Y = copy.deepcopy(X)
for i in range(X.shape[0]):
    Y[i,:] = X[i,:] + w # we are not assigning Y to new pointer, we just change its elements values


print(X) # so X does not change bec X and Y point to diffrent memory locations
print(Y)
```

```python
Y = X + w # broadcasting helps us to do this without possible problems 
# normally, their sizes are different so linear algebra does not allow this, but Python allows 
print(Y)
```

```python
w2 = [-1,0,1]
print(X * w2)
```

```python
w3 = [-1,0,1,2]
# print(X * w3) # gives error bec "operands could not be broadcast together with shapes (4,3) (4,)"
# we need to be sure about their size

w3 = np.array([-1,0,1,2], ndmin = 2) # minimum number of dimensions = 2 means 2D array
print(w3.shape) # we need to transpose

print(X * w3.T) # multp (4,3) with (4,1) - not dot product, this is like pointwise multp
# similar to addition
print(X + w3.T)

print(X / w3.T)
```
